#### objectif : Reproduction de l'Expérience DIORES : Prédiction de la Réussite Universitaire par Divers Algorithmes de Machine Learning à l'UCAD
   

```python
condition = (data['ANNEE UNIVERSITAIRE'] == '2018-2019') & (data['ETABLISSMENT_CODE'] =="FST") & (data['NIVEAU LMD ET NON LMD'].isin(['L1', '1A'])) 
data_filtre = data[condition]
```

```python
data_filtre
```

```python
data_filtre['RESULTAT'].value_counts()
```

```python
#prendre les étudiants qui ont réussi et redoubler
data = data_filtre[data_filtre['RESULTAT'].isin(['Passe', 'Redouble'])]
data['RESULTAT'].value_counts()
```

```python
data.isnull().sum()
```

```python
#suppresion les lignes avec des valeurs manquantes
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
#creer une nouvelle colonne age 
data["DATE_DE_NAISSANCE"] = pd.to_datetime(data["DATE_DE_NAISSANCE"], format='%d/%m/%Y', errors='coerce')
data['ANNEE DE NAISSANCE'] = data['DATE_DE_NAISSANCE'].dt.year
data['AGE'] = 2019 - data['ANNEE DE NAISSANCE']
data['AGE'].unique()
```

```python
data.columns
```

```python
#choix des colonnes a supprimer pour les modeles a venir 
colonnes_a_supprimer = [
    'NUMERO', 'INE', 'NOM', 'PRENOM', 'ETABLISSMENT_CODE',
    'ANNEE_INSCRIPTION', 'ANNEE UNIVERSITAIRE', 'CODE_NIVEAU',
    'DATE_DE_NAISSANCE', 'ANNEE DE NAISSANCE' , 'NATIONALITE', 'ANNEE_BACC'
]

data = data.drop(columns=colonnes_a_supprimer) 
```

```python
#preparation des variables x et y
x = data[data.columns.difference(['RESULTAT'])] 
y = data['RESULTAT']

#variable cible 
y = y.map({'Passe': 1, 'Redouble': 0})  

```

```python
data['RESULTAT'].value_counts() * 100 / len(data)
```

```python
#afficher les variables catégorielles
variables_categorielles = x.select_dtypes(include=['object']).columns.tolist()
variables_categorielles
```

```python
#encoder les variables catégorielles
x = pd.get_dummies(x, columns=variables_categorielles, drop_first=True)

```

```python
#les variables numériques
variables_numeriques = x.select_dtypes(include=['int', 'float64']).columns.tolist()
variables_numeriques
#normaliser les variables numériques
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x[variables_numeriques] = scaler.fit_transform(x[variables_numeriques])
#afficher les variables après normalisation
x[variables_numeriques].head()

```

```python
x.head()
```

```python
#transformer les bool en entier
for col in x.columns:
    if x[col].dtype == bool:
        x[col] = x[col].astype(int)
 
```

```python
x.head()
```

```python
x.describe()
#pour les variables nu meriques je dois avoir mean=0 et std=1  
#afficher les variables après normalisation
x[variables_numeriques].describe()
```

#### régression logistique

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
#separer le jeux en train et test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
regresiion_logistique = LogisticRegression()
regresiion_logistique.fit(x_train, y_train)
```

```python
#faire des predictions
regresiion_logistique_pred = regresiion_logistique.predict(x_test) 
```

```python
#calculer la précision 
score_regresiion_logistique = accuracy_score(y_test, regresiion_logistique_pred)
print(f"Précision de la régression logistique : {score_regresiion_logistique:.2f}") 
```

#### SVM(Support Vector Machine)

```python
#modele svm  
from sklearn.svm import SVC
model_svm = SVC() 
model_svm.fit(x_train, y_train)
```

```python
#faire des predictions
svm_pred = model_svm.predict(x_test)
```

```python
#calcul de la précision
score_svm = accuracy_score(y_test, svm_pred) 
print(f"Précision du modèle SVM : {score_svm:.2f}")
```

#### arbre de decision

```python
from sklearn.tree import DecisionTreeClassifier
arbre = DecisionTreeClassifier(random_state=42) 
arbre.fit(x_train, y_train)
```

```python
arbre_pred = arbre.predict(x_test)
```

```python
#calculer la précision
precision_arbre = accuracy_score(y_test, arbre_pred) 
print(f"Précision de l'arbre de décision : {precision_arbre:.2f}")
```

#### Random forest

```python
from sklearn.ensemble import RandomForestClassifier 
model_random = RandomForestClassifier(random_state=42)
model_random.fit(x_train, y_train)
```

```python
#faire des predictions
random_pred = model_random.predict(x_test)
```

```python
#calculer la précision
precision_random = accuracy_score(y_test, random_pred)
print(f"Précision du modèle Random Forest : {precision_random:.2f}") 
```
